```python
#1  Plot a simple line graph using the values: x = [1, 2, 3, 4], y = [10, 20, 25, 30]

import matplotlib.pyplot as plt
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]
plt.plot(x, y)
plt.show()

#2.  Add a title and axis labels to the line graph in Question 1.

import matplotlib.pyplot as plt
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]
plt.plot(x, y)
plt.title("Simple Line Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()

#3. Plot a red dotted line for: x = [0, 1, 2, 3], y = [3, 6, 1, 8]
import matplotlib.pyplot as plt
x = [0, 1, 2, 3]
y = [3, 6, 1, 8]
plt.plot(x, y, 'r--')
plt.show()

#4.  Plot two lines in the same graph: x = [1, 2, 3, 4], y1 = [10, 20, 25, 30], y2 = [5, 15, 20, 25]

import matplotlib.pyplot as plt
x = [1, 2, 3, 4]
y1 = [10, 20, 25, 30]
y2 = [5, 15, 20, 25]
plt.plot(x, y1)
plt.plot(x, y2)
plt.show()

#5. Add a legend to the above plot to differentiate the two lines.
import matplotlib.pyplot as plt
x = [1, 2, 3, 4]
y1 = [10, 20, 25, 30]
y2 = [5, 15, 20, 25]
plt.plot(x, y1, label="Line 1")
plt.plot(x, y2, label="Line 2")
plt.legend()
plt.show()

#6. Plot a bar chart for: Fruits = ['Apple', 'Banana', 'Mango'], Quantity = [10, 20, 15]
import matplotlib.pyplot as plt
fruits = ['Apple', 'Banana', 'Mango']
quantity = [10, 20, 15]
plt.bar(fruits, quantity)
plt.show()

#7. Plot a horizontal bar chart for the same fruit and quantity data.
import matplotlib.pyplot as plt
fruits = ['Apple', 'Banana', 'Mango']
quantity = [10, 20, 15]
plt.barh(fruits, quantity)
plt.show()

#8. Create a pie chart with percentage labels: Subjects = ['Math', 'Science', 'English'], Values = [30, 40, 30]
import matplotlib.pyplot as plt
subjects = ['Math', 'Science', 'English']
values = [30, 40, 30]
plt.pie(values, labels=subjects, autopct='%1.1f%%')
plt.show()

#9. Create a scatter plot: x = [5, 7, 8, 7], y = [8, 5, 6, 3]
import matplotlib.pyplot as plt
x = [5, 7, 8, 7]
y = [8, 5, 6, 3]
plt.scatter(x, y)
plt.show()

#10. Create a histogram for the data: marks = [60, 70, 80, 75, 70, 90, 85, 80]
import matplotlib.pyplot as plt
marks = [60, 70, 80, 75, 70, 90, 85, 80]
plt.hist(marks, bins=5)
plt.show()

#11.  Create a subplot with two plots: one line plot and one bar chart.
import matplotlib.pyplot as plt
fig, (ax1, ax2) = plt.subplots(1, 2)
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]
ax1.plot(x, y)
ax1.set_title("Line Plot")
fruits = ['Apple', 'Banana', 'Mango']
quantity = [10, 20, 15]
ax2.bar(fruits, quantity)
ax2.set_title("Bar Chart")
plt.tight_layout()
plt.show()

#12. Change the figure size to 8x5 while plotting a simple graph.
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 5))
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]
plt.plot(x, y)
plt.show()

#13. Plot a graph with customized line style, marker, and color.
import matplotlib.pyplot as plt
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]
plt.plot(x, y, color='green', linestyle='--', marker='o')
plt.show()

#14.  Save a plot as a PNG file using plt.savefig()
import matplotlib.pyplot as plt
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]
plt.plot(x, y)
plt.savefig('plot.png')
plt.show()

#15.  Add grid lines to any plot you create.
import matplotlib.pyplot as plt
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]
plt.plot(x, y)
plt.grid(True)
plt.show()

#16.  Plot a line graph and annotate the highest point on it.
import matplotlib.pyplot as plt
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]
plt.plot(x, y)
max_y = max(y)
max_x = x[y.index(max_y)]
plt.annotate(f'Highest: {max_y}', xy=(max_x, max_y), xytext=(max_x, max_y+2))
plt.show()

#17.  Create a bar chart and add value labels on top of each bar.
import matplotlib.pyplot as plt
fruits = ['Apple', 'Banana', 'Mango']
quantity = [10, 20, 15]
bars = plt.bar(fruits, quantity)
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.5, yval, ha='center')
plt.show()

#18. Create a plot with x-axis rotated 45 degrees using tick labels: Months = ['Jan', 'Feb', 'Mar', 'Apr'], Sales = [250, 300, 150, 400]
import matplotlib.pyplot as plt
months = ['Jan', 'Feb', 'Mar', 'Apr']
sales = [250, 300, 150, 400]
plt.bar(months, sales)
plt.xticks(rotation=45)
plt.show()

#19. Draw three subplots in a single row, each with a different plot (line, bar, pie).
import matplotlib.pyplot as plt
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(12, 4))
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]
ax1.plot(x, y)
ax1.set_title("Line Plot")
fruits = ['Apple', 'Banana', 'Mango']
quantity = [10, 20, 15]
ax2.bar(fruits, quantity)
ax2.set_title("Bar Chart")
subjects = ['Math', 'Science', 'English']
values = [30, 40, 30]
ax3.pie(values, labels=subjects, autopct='%1.1f%%')
ax3.set_title("Pie Chart")
plt.tight_layout()
plt.show()

#20. Use different colors for each bar in a bar chart using the color parameter.
import matplotlib.pyplot as plt
fruits = ['Apple', 'Banana', 'Mango']
quantity = [10, 20, 15]
plt.bar(fruits, quantity, color=['red', 'yellow', 'green'])
plt.show()

```
