10. Use the 'titanic' dataset and show a heatmap of its numeric correlation matrix.

```python
df1 = sns.load_dataset('titanic')
df1.head()
```

```python
corr=df1.corr(numeric_only=True)
sns.heatmap(corr,annot=True,cmap='viridis')
plt.title("Heatmap",fontsize=15)
plt.show()
```

18. Convert all categorical columns in the 'tips' dataset to numeric and create a full correlation heatmap.

```python
print(df['day'].unique())
print(df['time'].unique())
```

```python
df=pd.get_dummies(df,columns=['sex','smoker','time'],drop_first=True,dtype=int)
df.head()
```
